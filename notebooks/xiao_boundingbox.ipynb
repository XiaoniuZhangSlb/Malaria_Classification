```python
import cv2
import os

def generate_boudningbox(image_path, output_path, label="1"):
    # Read the image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Threshold the image to identify white spots
    _, thresh = cv2.threshold(gray, 230, 255, cv2.THRESH_BINARY)

    # Find contours of white spots
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Write normalized coordinates to a file in YOLO format
    height, width, _ = image.shape
    with open(output_path, 'w') as file:
        for contour in contours:
            # Get bounding box coordinates
            x, y, w, h = cv2.boundingRect(contour)

            # Normalize coordinates
            x_center = (x + w / 2) / width
            y_center = (y + h / 2) / height
            normalized_width = w / width
            normalized_height = h / height

            # Write to YOLO format
            line = f"{label} {x_center:.6f} {y_center:.6f} {normalized_width:.6f} {normalized_height:.6f}\n"
            file.write(line)

            # Draw bounding box on the image
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

    output_image_path = os.path.join(f"{os.path.dirname(output_path)}", f"{os.path.basename(image_path)}")

    # Save the image with bounding boxes
    cv2.imwrite(output_image_path, image)
```

```python
import os
import random
import shutil

def split_data(input_folder, output_train, output_valid, output_test, input_label_folder, train_ratio=0.6, valid_ratio=0.2, test_ratio=0.2, file_extension=".jpg"):
    # List all files in the input folder with the specified extension
    all_files = [file for file in os.listdir(input_folder) if file.endswith(file_extension)]
    num_files = len(all_files)

    # Shuffle the list of files
    random.shuffle(all_files)

    # Calculate the number of files for each set
    num_train = int(train_ratio * num_files)
    num_valid = int(valid_ratio * num_files)
    num_test = num_files - num_train - num_valid

    # Split files into train, valid, and test sets
    train_files = all_files[:num_train]
    valid_files = all_files[num_train:num_train + num_valid]
    test_files = all_files[num_train + num_valid:]

    # Copy files to output folders
    for file in train_files:
        shutil.copy(os.path.join(input_folder, file), os.path.join(output_train, 'images', file))
        label_file = os.path.join(input_label_folder, f"{os.path.splitext(file)[0]}.txt")
        if os.path.exists(label_file):
            shutil.copy(label_file, os.path.join(output_train, 'labels', f"{os.path.splitext(file)[0]}.txt"))
        else:
            print(f"The file {label_file} does not exist.")
    for file in valid_files:
        shutil.copy(os.path.join(input_folder, file), os.path.join(output_valid, 'images',file))
        label_file = os.path.join(input_label_folder, f"{os.path.splitext(file)[0]}.txt")
        # print(os.path.join(output_valid, 'labels', f"{os.path.splitext(file)[0]}.txt"))
        if os.path.exists(label_file):
            shutil.copy(label_file, os.path.join(output_valid, 'labels', f"{os.path.splitext(file)[0]}.txt"))
        else:
            print(f"The file {label_file} does not exist.")
    for file in test_files:
        shutil.copy(os.path.join(input_folder, file), os.path.join(output_test,'images', file))
        label_file = os.path.join(input_label_folder, f"{os.path.splitext(file)[0]}.txt")
        if os.path.exists(label_file):
            shutil.copy(label_file, os.path.join(output_test, 'labels', f"{os.path.splitext(file)[0]}.txt"))
        else:
            print(f"The file {label_file} does not exist.")
```

```python
import cv2
import numpy as np

def draw_bounding_boxes(image_path, output_file):
    # Read the image
    image = cv2.imread(image_path)

    # Convert the image to grayscale for contour detection
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to detect regions
    _, thresh = cv2.threshold(gray, 230, 255, cv2.THRESH_BINARY)

    # Find contours in the binary image
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Class information and colors
    class_colors = {'parasite': (0, 255, 0), 'background': (0, 0, 255)}

    # Open the output file for writing coordinates
    with open(output_file, 'w') as f:
        for contour in contours:
            # Get the bounding box coordinates
            x, y, w, h = cv2.boundingRect(contour)

            # Classify objects based on position, size, or other criteria
            # class_label = 'parasite' if w > 10 and h > 10 else 'background'
            class_label = 'normal'
            # Draw the bounding box on the original image with class-specific color
            color = class_colors.get(class_label, (255, 255, 255))
            cv2.rectangle(image, (x, y), (x + w, y + h), color, 2)

            # Write the coordinates and class label to the output file
            f.write(f'{class_label} {x + w/2} {y + h/2} {w} {h}\n')

    # Display the image with bounding boxes
    cv2.imshow('Image with Bounding Boxes', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Specify the path to your image and output file
image_path = '/home/xiao/code/XiaoniuZhangSlb/Malaria_Classification/raw_data/MP-IDB-The-Malaria-Parasite-Image-Database-for-Image-Processing-and-Analysis-master/Malariae/img/1312132815-0001-G.jpg'
output_file = 'output_coordinates.txt'

# Call the function to draw bounding boxes and label different classes
draw_bounding_boxes(image_path, output_file)
```
